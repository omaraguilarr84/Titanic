```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = "/Users/omaraguilarjr/Library/CloudStorage/OneDrive-GeorgiaInstituteofTechnology/Kaggle/titanic/"

train_df = pd.read_csv(data_path + "train.csv")
test_df = pd.read_csv(data_path + "test.csv")
```

```python
print(train_df[:10])
```

```python
sns.countplot(x="Survived", data=train_df, palette='coolwarm')
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Sex", data=train_df, palette="coolwarm")
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=train_df, palette="coolwarm")
plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(train_df["Age"].dropna(), bins=30, kde=True, color="blue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(train_df["Fare"], bins=30, kde=True, color="green")
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
sns.pairplot(train_df[["Survived", "Pclass", "Age", "Fare", "Sex", "SibSp", "Parch", "Embarked"]], hue="Survived", palette="coolwarm")
plt.show()
```
